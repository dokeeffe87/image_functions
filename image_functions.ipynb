# Collecting some potentially useful image functions

```python
import os 
import sys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load test images:
image_bay = plt.imread('Unequalized_Hawkes_Bay_NZ.jpg')
image_nebula = plt.imread('hs-2004-32-b-small_web.jpg')
```

```python
def convert_colour_to_grayscale(img, image_name, colorbar=False):
    # Compute the intensity of the image a the sum of the red, green and blue channels
    intensity = img.sum(axis=2)
    plt.imshow(intensity, cmap='gray')
    if colorbar:
        plt.colorbar()
    # Hide the axes and show the figure
    plt.axis('off')
    plt.show()
    # Save the figure
    plt.imsave(image_name, intensity, cmap='gray')
```

```python
plt.imshow(image_nebula)
```

```python
# Convert this to black and white.
convert_colour_to_grayscale(image_nebula, 'nebula_grayscale.jpg', colorbar=False)
```
